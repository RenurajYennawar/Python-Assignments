## 3. Indexing and Slicing

### 3.1 Indexing

The items of an array can be accessed and assigned to the same way as other Python sequences (e.g. lists):


```python
a = np.arange(10)
print(a)
print(a[5])  #indices begin at 0, like other Python sequences (and C/C++)
```

```python
b = np.diag([3,4,5])
print(b)

print(b[0][0])
```

```python
c = np.array([  [   [0, 1], [2, 3]   ], [  [4, 5], [6, 7]  ]   ])
print(c)
print(c[0][1][1])
```

### 3.2 Slicing

```python
array = np.arange(10)
print(array)
print(array[0:10:2])
```

```python
array[5:7] = 90
print(array)
```

```python
b = np.arange(5)
array[0:5] = b[::-1]
print(array)
```

```python
b = np.diag([3,4,5])
print(b)
print("*"*10)
b[2][1:] = 99
print(b)
```

### 4. Copies and Views

1. A slicing operation creates a view on the original array, which is just a way of accessing array data. 
2. Thus the original array is not copied in memory. 
3. You can use np.may_share_memory ( ) to check if two arrays share the same memory block.
4. When modifying the view, the original array is modified as well:

```python
arrA = np.arange(10)

arrB = arrA[::2] # arrB also points to arrA by sharing memory
print("arrA: ",arrA)
print("arrB: ",arrB)

# print("Address of arrA: ",id(arrA))
# print("Address of arrB: ",id(arrB))
# print("\n arrA & arrB shares Memory? ", np.shares_memory(arrA,arrB))
arrA[4] = 999

print("\n arrA: ",arrA) # Both arrA and arrB shares same memory as data is allocated using slicing
print(" arrB: ",arrB) 

```

```python
# To force a copy  we have to use copy() method from NumPy
arrA = np.arange(10)

arrB = arrA[::2].copy()
print("arrA: ",arrA)
print("arrB: ",arrB)

print("Address of arrA: ",id(arrA))
print("Address of arrB: ",id(arrB))
print("\n arrA & arrB shares Memory? ", np.shares_memory(arrA,arrB))
arrB[4] = 999

print("\n arrA: ",arrA) # Both arrA and arrB do not shares same memory as data is allocated using copy()
print(" arrB: ",arrB) 
```

### 5. Fancy Indexing

```python
arr = np.arange(5)
print(arr)
mask = (arr%2 == 0)
print(mask)

arrB = arr[mask] # mask statement is "mapped" to arr and whenever the statement is true then only data is
                 # allocated to arrB

print("arr: ",arr)
print("arrB: ",arrB)
```

```python
arr[mask] = -1 # all places in arr where even numbers were there replaced with -1
print(arr)
```
